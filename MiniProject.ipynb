# Exploratory Data Analysis of ICC World Cup Matches


```python
#importing libraries
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#Reading datasets(.csv files)
players = pd.read_csv('WC_players.csv')
ground_data = pd.read_csv('Ground_Averages.csv')
batsman = pd.read_csv('Batsman_Data.csv')
bowler = pd.read_csv('Bowler_data.csv')
odi_total = pd.read_csv('ODI_Match_Totals.csv')
odi_total = pd.read_csv('ODI_Match_Totals.csv')
```

## Data Analysis of Indian Team

```python
#Creating Dataframe 'India'
India = odi_total[odi_total['Country'] == 'India']

def isWin(val):
  if val in ['won','lost']:
    if val ==  'won':
      color = 'green'
    elif val == 'lost':
      color = 'red'
    else:
      color = 'yellow'
  else:
    color = 'black'
  return 'color: %s' % color
India.style.applymap(isWin)
```

```python
# Finding null values
India.isnull()
```

```python
#Droping null values
India = India.dropna()

#Displaying DataFrame
India
```

```python
India = odi_total[odi_total['Country'] == 'India']

listInd = []
for item in India['RPO'].tolist():
  if item == '-':
    continue
  listInd.append(float(item))

plt.ylabel('RPO')
plt.xlabel('Match Number')
plt.plot(range(0,155),listInd)
plt.title('RPO trend in Different matches')
plt.show()
```

```python
#displaying matches which India had won.
India_win = India[India['Result'] == 'won']
India_win
```

```python
#Displaying opposition against whom India has won.
India_win['Opposition'].value_counts()
```

```python
#plotting bar chart of matches won by india
plt.figure(figsize=(8,5))
plt.bar(list(India_win['Opposition'].value_counts()[0:5].keys()), list(India_win['Opposition'].value_counts()[0:5]), color="g")
plt.title('Wins of INDIA')
plt.xlabel('Opponent Country')
plt.ylabel('Number of matches')
plt.show() 
```

```python
#Plotting a pie chart of matches Won by India
plt.figure(figsize=(7,7))
plt.pie(list(India_win['Opposition'].value_counts()),labels=list(India_win['Opposition'].value_counts().keys()),autopct='%0.1f%%')
plt.legend(labels=list(India_win['Opposition'].value_counts().keys()))
plt.title('Percentage of Matches INDIA Won Against different countries')
plt.show()
```

```python
#displaying matches which India had lost.
India_loss = India[India['Result']=='lost']
#dropping null values
India_loss = India_loss.dropna()

#Displaying Dataframe
India_loss.head()
```

```python
#Displaying opposition against whom India has lost.
India_loss['Opposition'].value_counts()
```

```python
#plotting bar chart of matches lost by India.
plt.figure(figsize=(8,5))
plt.bar(list(India_loss['Opposition'].value_counts()[0:3].keys()),list(India_loss['Opposition'].value_counts()[0:3]),color="g")
plt.title('Lost of INDIA')
plt.xlabel('Opponent Country')
plt.ylabel('Number of matches')
plt.show() 
```

```python
#Plotting a pie chart of Matches lost by India
plt.figure(figsize=(7,7))
plt.pie(list(India_loss['Opposition'].value_counts()),labels=list(India_loss['Opposition'].value_counts().keys()),autopct='%0.1f%%')
plt.legend(labels=list(India_loss['Opposition'].value_counts().keys()))
plt.title('Percentage of Matches INDIA lost Against different countries')
plt.show()
```
